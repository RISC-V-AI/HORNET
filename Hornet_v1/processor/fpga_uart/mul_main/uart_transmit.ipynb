```python
file_x=open("output_padding.txt", "r")
for i in range (50):
    data = file_x.readline()
    data = int(data, 16)
    length = (data.bit_length() + 7) // 8
    b = data.to_bytes(length, byteorder='little')
    
    bytes2send = b.ljust(59, b'\x00')
    
    print(f"{bytes2send.hex()}")
    
    ser.write(bytes2send) 
    #ser.write(b'\n') 
    received_data = ser.read(59)
    print(f"{received_data.hex()}")

   
file_x.close()    
```
